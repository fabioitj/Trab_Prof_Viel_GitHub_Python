## Matplotlib

Aqui iremos importar a biblioteca "matplotlib" e "numPy"


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Abaixo nós iremos definir o valor de X

```python
x = np.linspace(0,1,10)
```

Primeiramente iremos definir a primeira linha "Quadrado" com a cor verde, o x no gráfico começará em 0 e irá recebendo seu próprio quadrado (Ex: 1 2 4 8 ...)

```python
plt.plot(x, x**2, 'g.-', label='quadrado')
plt.plot(x, np.sqrt(x) / 3, 'r.-', label='raiz quadrada')
plt.xlabel('x');
plt.ylabel('y');
plt.title('Título');
plt.legend();
plt.show()
```
